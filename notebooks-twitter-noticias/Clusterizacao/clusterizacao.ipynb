```python
df_teste = tweets_content.iloc[:12000]
```

```python
df_teste['clean_text'] = hero.clean(df_teste['text'])
```

```python
hero.tfidf(df_teste['clean_text'])
```

```python
tweets = pd.Series(df_teste['clean_text']) 
hero.term_frequency (tweets)
```

```python
column_names = ["text","tfidf", "kmeans_labels"]
```

```python
tweets = pd.DataFrame (columns=column_names)
```

```python
tweets["text"] = df_teste.clean_text
```

```python
tweets['tfidf'] = (
    tweets['text']
    .pipe(hero.tfidf)
)
```

```python
tweets['kmeans_labels'] = (
    tweets['tfidf']
    .pipe(hero.kmeans, n_clusters=5)
    .astype(str)
)
```

```python
tweets[["text","kmeans_labels"]].head(10)
```

```python
tweets['pca'] = tweets['tfidf'].pipe(hero.pca)
```

```python
hero.scatterplot(tweets, 'pca', color="kmeans_labels", title="Tweets")
```

```python
tweets.to_csv('kmeans-labels-df3')
```
